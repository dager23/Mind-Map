```python
#OLD Method - Not efficient

def set_threshold(phrases):
    total_score = 0
    max_score = 0
    min_score = math.inf
    for phrase in phrases:
        score = phrase[0]
        if (score > max_score):
            max_score = score
        if(score < min_score):
            min_score = score
        total_score += score
    mean_score = total_score / len(phrase)
    print(total_score, max_score, min_score, mean_score)
    return (total_score /  abs(max_score - mean_score))
```

```python
def set_threshold_quartile(phrases):
    scores = []
    for phrase in phrases:
        scores.append(phrase[0])
    threshold = np.percentile(scores, 75)
    return threshold
```

```python
def set_threshold_variance(phrases):
    scores = []
    for phrase in phrases:
        scores.append(phrase[0])
    std_dev = np.var(scores) ** 0.5
    total = sum(scores)
    mean_scores = total / len(scores)
    threshold = mean_scores + std_dev * 2
    print(threshold)
    return threshold
```

```python
def get_best_phrases(phrases, threshold=4):
    final_phrases = []
    for phrase in phrases:
        if(phrase[0] > math.floor(threshold)):
            final_phrases.append(phrase[1])

    return final_phrases
```

```python
def get_best_phrases(phrases, threshold=4):
    final_phrases = []
    for phrase in phrases:
        if(phrase[0] > math.floor(threshold)):
            if(final_phrases):
                flag = 0
                for prev_phrase in final_phrases:
                    similarity = mm.sentence_similarity(prev_phrase, phrase[1], stop_words)
                    if(similarity > 0.96):
                        flag = 1
                        break
                if(not flag):
                    final_phrases.append(phrase[1])
            else:
                final_phrases.append(phrase[1])

    return final_phrases
```

```python
def get_keywords(grouped_text):
    final_keywords = []
    stop_words = stopwords.words('english')
    for idx in grouped_text:
        if(grouped_text[idx]):
            rake_model = Rake()
            rake_model.extract_keywords_from_text(grouped_text[idx])
            phrases = rake_model.get_ranked_phrases_with_scores()
            threshold = set_threshold_quartile(phrases)
            best_phrases = get_best_phrases(phrases, threshold)
            final_keywords.append(best_phrases)
        else:
            final_keywords.append([])
    return final_keywords
```
